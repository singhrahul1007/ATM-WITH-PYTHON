```python
from tkinter import*
from tkinter import messagebox
mainMenu = Tk()
mainMenu.geometry('500x500')
mainMenu.title(" Main page")
#login_function
def open():
    top=Toplevel()
    top.geometry('500x500')
    top.title("LOGIN PAGE")
    label_0=Label(top,text="Email",width=20,font=("bold",10))
    label_0.place(x=90,y=203)
    entry_0 = Entry(top)
    entry_0.place(x=240,y=203)
    label_1=Label(top,text="FullName:",width=20,font=("bold",10))
    label_1.place(x=90,y=255)
    entry_1 = Entry(top)
    entry_1.place(x=240,y=255)
    var = IntVar()
    Radiobutton(top, text="Remmember me",padx = 5, variable=var, value=1).place(x=225,y=300)
    Button(top, text='login',width=20,bg='blue',fg='white', command=pop_1).place(x=180,y=350)
    label_2=Label(top,text="If don't have login id the click on the registration",width=40,font=("bold",10))
    label_2.place(x=90,y=375)
    Button(top, text='Registration',width=20,bg='blue',fg='white', command=reg).place(x=180,y=400)



#registration window
def reg():
    root = Tk()

    root.geometry('500x500')
    root.title("Registration Form")

label_0 = Label(root, text="Registration form",width=20,font=("bold", 20))
label_0.place(x=90,y=53)


label_1 = Label(root, text="FullName",width=20,font=("bold", 10))
label_1.place(x=80,y=130)

 entry_1 = Entry(root)
 entry_1.place(x=240,y=130)

 label_2 = Label(root, text="Email",width=20,font=("bold", 10))
 label_2.place(x=68,y=180)

 entry_2 = Entry(root)
 entry_2.place(x=240,y=180)

 label_3 = Label(root, text="Gender",width=20,font=("bold", 10))
 label_3.place(x=70,y=230)
 var = IntVar()
 Radiobutton(root, text="Male",padx = 5, variable=var, value=1).place(x=235,y=230)
 Radiobutton(root, text="Female",padx = 20, variable=var, value=2).place(x=290,y=230)

 label_4 = Label(root, text="Age:",width=20,font=("bold", 10))
 label_4.place(x=70,y=280)


 entry_2 = Entry(root)
 entry_2.place(x=240,y=280)


 Button(root, text='Submit',width=20,bg='brown',fg='white', command=pop_0).place(x=180,y=380)


 label_4 = Label(root, text="If you have a login id then please click on the login.",width=40,font=("bold", 10))
 label_4.place(x=100,y=420)

 Button(root, text='LOGIN',width=20,bg='blue',fg='white', command=open).place(x=180,y=450)

# main window
def pop_0():
  messagebox.showinfo("sign up", "you have been registered")
def pop_1():
    messagebox.showinfo("login", "welcome back")


   
label_4 = Label(mainMenu, text="Choose your option:",width=40,font=("bold", 15))
label_4.place(x=40,y=80)
Button(mainMenu, text='login',width=20,bg='blue',fg='white', command=open).place(x=180,y=150)
Button(mainMenu, text='Registration',width=20,bg='blue',fg='white', command=reg).place(x=180,y=200)
Button(mainMenu, text='Quit',width=20,bg='Red',fg='white', command=mainMenu.destroy).place(x=180,y=250)

mainMenu.mainloop()
print("registration form  seccussfully created...")
```
